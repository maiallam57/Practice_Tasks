Use perspective transform to crop the subway car face from the given image. In this task you must
create a functional GUI where the user can delete wrong points and can reattempt the cropping process.

```python
import cv2
import numpy as np

subwaycar = cv2.imread('D:/vortex/Codes/Task 3/photomosaic-test-1c_51268761941_o.png')
subwaycar = cv2.resize(subwaycar, (450,250))
h, w, c = subwaycar.shape
circles = []
temp = subwaycar.copy()

def Sort(Input): 
    pos1_0, pos1_1 = w/2, h/2
    pos2_0, pos2_1 = w/2, h/2
    pos3_0, pos3_1 = w/2, h/2
    pos4_0, pos4_1 = w/2, h/2

    lst = len(Input)
    for i in range(lst):
        for j in range(2):
            current_sum = Input[j][0]+Input[j][1]

            if current_sum < (pos1_0 + pos1_1) and current_sum < (pos4_0 + pos4_1) and current_sum < (pos3_0 + pos3_1)and current_sum < (pos2_0 + pos2_1):
                pos1_0, pos1_1 = Input[j][0], Input[j][1]

            if current_sum > (pos4_0 + pos4_1) and current_sum > (pos1_0 + pos1_1)and current_sum > (pos3_0 + pos3_1)and current_sum > (pos2_0 + pos2_1):
                pos4_0, pos4_1 = Input[j][0], Input[j][1] 

            current_sub = Input[j][0]-Input[j][1]
            if abs(current_sub) > abs(pos2_0 - pos2_1) and abs(current_sub) > abs(pos3_0 - pos3_1) and abs(current_sub) > abs(pos1_0 - pos1_1) and abs(current_sum) > abs(pos4_0 - pos4_1):
                pos2_0, pos2_1 = Input[j][0], Input[j][1]

            if abs(current_sub) < abs(pos3_0 - pos3_1) and abs(current_sub) < abs(pos2_0 - pos2_1) and abs(current_sub ) < abs(pos1_0 - pos1_1) and abs(current_sum) < abs(pos4_0 - pos4_1):
                pos3_0, pos3_1 = Input[j][0], Input[j][1]
                
        print("p1= ( " + str(pos1_0) +","+ str(pos1_1)+")" )
        print("p4= ( " + str(pos4_0) +","+ str(pos4_1)+")" )
        print("p2= ( " + str(pos2_0) +","+ str(pos2_1)+")" )
        print("p3= ( " + str(pos3_0) +","+ str(pos3_1)+")" )

    points =[(pos1_0, pos1_1), (pos2_0, pos2_1), (pos3_0, pos3_1), (pos4_0, pos4_1)]
    print("points= ",points)
    return points 


def draw_circle(event, x, y, flags, param):

    global circles
    if event == cv2.EVENT_LBUTTONDOWN: # Add the circle
        circles.append((x, y))

    if event == cv2.EVENT_RBUTTONDBLCLK: # Delete all circles
        circles[:] = []

    elif event == cv2.EVENT_RBUTTONDOWN: # Delete last added circle
        try:
            circles.pop()
        except (IndexError):
            print("no circles to delete")

cv2.namedWindow('subway car')

cv2.setMouseCallback('subway car', draw_circle)

dst_image = clone = subwaycar.copy()


while True:
    image = clone.copy()

    

    for pos in circles:
        cv2.circle(image, pos, 5, (255, 0, 0), 3)

    im_v = cv2.vconcat([image, dst_image])

    cv2.imshow("subway car", im_v)

    
    try:
        pts_1 = np.float32(circles)
        pts_2 = np.float32([[0, 0], [w, 0], [0, h], [w, h]])
        M = cv2.getPerspectiveTransform(pts_1, pts_2)

        dst_image = cv2.warpPerspective(temp, M, (w, h))
    except:
        pass

    key = cv2.waitKey(1)
    if key == ord('q') or key == ord("Q"):
        cv2.destroyAllWindows()
        break

print(circles)
circles = Sort(circles)

```
