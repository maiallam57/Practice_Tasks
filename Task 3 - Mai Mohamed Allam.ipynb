TASK 3: 
Use perspective transform to crop the subway car face from the given image. In this task you must
create a functional GUI where the user can delete wrong points and can reattempt the cropping process.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


subwaycar = cv2.imread('Assignment/Task 3/photomosaic-test-1c_51268761941_o.png')
h, w, c = subwaycar.shape
circles = []

temp = subwaycar.copy()

menu_pos5 = (10, 550)

def draw_text():
    menu_pos = (10, 475)
    menu_pos2 = (10, 500)
    menu_pos3 = (10, 525)
    menu_pos4 = (10, 550)
   

    cv2.putText(subwaycar, 'Simple left click: Add a circle', menu_pos, cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 0))
    cv2.putText(subwaycar, 'Simple right click: Delete last circle', menu_pos2, cv2.FONT_HERSHEY_SIMPLEX, 0.8,(0, 0, 0))
    cv2.putText(subwaycar, 'Double right click: Delete all circle', menu_pos3, cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 0))
    cv2.putText(subwaycar, 'Press \'q\' to exit', menu_pos4, cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 0))


def draw_circle(event, x, y, flags, param):
    global circles
    if event == cv2.EVENT_LBUTTONDOWN: # Add the circle
        circles.append((x, y))

    if event == cv2.EVENT_RBUTTONDBLCLK: # Delete all circles
        circles[:] = []

    elif event == cv2.EVENT_RBUTTONDOWN: # Delete last added circle
        try:
            circles.pop()
        except (IndexError):
            print("no circles to delete")


def repeat(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDBLCLK: # reattempt the cropping process
        circles[:] = []
        program()


def program():
    cv2.namedWindow('subway car')

    cv2.setMouseCallback('subway car', draw_circle)

    draw_text()

    clone = subwaycar.copy()

    while True:
        image = clone.copy()

        for pos in circles:
            cv2.circle(image, pos, 5, (255, 0, 0), 3)

        cv2.imshow('subway car', image)

        key = cv2.waitKey(1)
        if key == ord('q') or key == ord("Q"):
            break


    try:
        pts_1 = np.float32(circles)
        pts_2 = np.float32([[0, 0], [w, 0], [0, h], [w, h]])
        M = cv2.getPerspectiveTransform(pts_1, pts_2)

        dst_image = cv2.warpPerspective(temp, M, (w, h))

        cv2.putText(dst_image, 'Double left click: to crop again and go the original image', menu_pos5, cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 0))

        cv2.imshow("after perespective transform", dst_image)
    except:
        print(" ")


program()

cv2.setMouseCallback('after perespective transform', repeat)

key = cv2.waitKey(0)
cv2.destroyAllWindows()
```
