```python
import cv2
import matplotlib.pyplot as plt

def show_with_matplotlib(color_img, title, pos):

    img_RGB = color_img[:, :, ::-1]
    ax = plt.subplot(6, 4, pos)
    plt.imshow(img_RGB)
    plt.title(title)
    plt.axis('off')

image = cv2.imread('C:/Users/Mai_Allam/Desktop/opencv/assets/images/rgb.png')
plt.figure(figsize=(15, 15))
plt.suptitle('ColorSpaces', fontsize=14, fontweight='bold')

show_with_matplotlib(image, "RGB image",1)

B, G, R = cv2.split(image)

show_with_matplotlib(cv2.cvtColor(B,cv2.COLOR_GRAY2BGR),"BGR (B)",2)
show_with_matplotlib(cv2.cvtColor(G,cv2.COLOR_GRAY2BGR),"BGR (G)",3)
show_with_matplotlib(cv2.cvtColor(R,cv2.COLOR_GRAY2BGR),"BGR (R)",4)

image_copy = cv2.merge((B,G,R))

show_with_matplotlib(image_copy, "BGR image-copy",5)

b_copy = image[:, :, 0]

img_without_blue = image.copy()
img_without_blue[:,:,0] = 0

img_without_green = image.copy()
img_without_green[:,:,1] = 0

img_without_red = image.copy()
img_without_red[:,:,2] = 0

show_with_matplotlib(img_without_blue,"BGR without(B)",6)
show_with_matplotlib(img_without_green,"BGR without(G)",7)
show_with_matplotlib(img_without_red,"BGR without(R)",8)

B, G, R = cv2.split(img_without_blue)
show_with_matplotlib(cv2.cvtColor(B,cv2.COLOR_GRAY2BGR),"BGR without B (B)",9)
show_with_matplotlib(cv2.cvtColor(G,cv2.COLOR_GRAY2BGR),"BGR without B (G)",10)
show_with_matplotlib(cv2.cvtColor(R,cv2.COLOR_GRAY2BGR),"BGR without B (R)",11)

B, G, R = cv2.split(img_without_green)
show_with_matplotlib(cv2.cvtColor(B,cv2.COLOR_GRAY2BGR),"BGR without G(B)",13)
show_with_matplotlib(cv2.cvtColor(G,cv2.COLOR_GRAY2BGR),"BGR without G(G)",14)
show_with_matplotlib(cv2.cvtColor(R,cv2.COLOR_GRAY2BGR),"BGR without G(R)",15)

B, G, R = cv2.split(img_without_red)
show_with_matplotlib(cv2.cvtColor(B,cv2.COLOR_GRAY2BGR),"BGR without R(B)",17)
show_with_matplotlib(cv2.cvtColor(G,cv2.COLOR_GRAY2BGR),"BGR without R(G)",18)
show_with_matplotlib(cv2.cvtColor(R,cv2.COLOR_GRAY2BGR),"BGR without R(R)",19)

plt.show()
```
